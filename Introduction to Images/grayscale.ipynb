# 5. Converting RGB to HSV

```python
# there is the manual implementation, but it doesn't make sense
# I'll use the openCV implementation of it.


im = cv2.imread(r'flowers_colored.jpg')
im = im[:,:,::-1]
HsvIm = cv2.cvtColor(im,cv2.COLOR_BGR2HSV)
```

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(im)
plt.subplot(122)
plt.imshow(HsvIm)
```

```python
im = cv2.imread(r'flowers_colored.jpg')
HsvIm = cv2.cvtColor(im,cv2.COLOR_BGR2HSV)

lr = np.array([0,120,70])
ur = np.array([15,255,255])
mask1 = cv2.inRange(HsvIm,lr,ur)

lr = np.array([165,120,70])
ur = np.array([180,255,255])
mask2 = cv2.inRange(HsvIm,lr,ur)

mask = mask1 | mask2
res = cv2.bitwise_and(im, im, mask=mask)

plt.figure(1)
plt.subplot(121)
plt.imshow(im[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
```
