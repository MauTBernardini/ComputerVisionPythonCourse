# 0. Imports

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import cvlib as cv 
from cvlib.object_detection import draw_bbox
import imutils
```

# 1. CV Detector

```python
im = cv2.imread('faces_image.jpg')
im = cv2.resize(im, (300, 300))
plt.imshow(im[:,:,::-1])
```

```python
bbox , conf = cv.detect_face(im)
# loop through detected faces and add bounding box
for face in bbox:
    (startX,startY) = face[0],face[1]
    (endX,endY) = face[2],face[3]
    # draw rectangle over face
    cv2.rectangle(im, (startX,startY), (endX,endY), (0,255,0), 2)
# display output        
plt.imshow(im[:,:,::-1])
plt.show()
```

# 2. YOLO

```python
im = cv2.imread('persons3.jpg')
im = cv2.resize(im, (300, 300))
plt.imshow(im[:,:,::-1])

bbox , labels, conf = cv.detect_common_objects(im,model='yolov3-tiny')
output_image = draw_bbox(im,bbox,labels,conf)
plt.imshow(output_image[:,:,::-1])
```
